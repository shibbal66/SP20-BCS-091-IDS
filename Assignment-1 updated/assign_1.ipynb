```python
#Importing all necessARY libraries
import pandas as pd
import re
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
#reading csv file
csvData= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/fa22.csv')
csvData
#Printing the list of all students whose first name starts with letter the 'H'.
name = csvData[csvData.Name.str.contains('^H')]
name.Name
```

```python
#Print the total number of students who have a three words name (first-middle-surname).
â€¢ Print
csvData[csvData['Name'].str.split(' ').str.len()==3]['Name'].count()
```

```python
#Printing the percentage of students who have a CGPA of 3.0 or above.
CGPA = csvData.CGPA

GPA=csvData[csvData.CGPA.str.contains('^3')]

getGPA = GPA.CGPA

totalS = 0
All= 0
for above in getGPA:
  totalS+=1
for total in CGPA:
    All+=1
    
percent = (totalS * 100) / All
print(percent)


```

```python
#Plotting a pie chart to show the ratio of male and female students.
male = csvData[csvData.Gender.str.contains('^M')]
get_male_stud = male.Gender

count_male =0

count_female=0
female = csvData[csvData.Gender.str.contains('^F')]
for x in get_male_stud:
   count_male +=1

get_female_stud = female.Gender
for y in get_female_stud:
   count_female +=1

y=np.array([count_male,count_female])
f = ['Male', 'Female']
plt.pie(y,labels=f,autopct= '%1.2f%%')
plt.legend()
plt.show()
```

```python
#Plotting the CGPA of all male students on a histogram with intervals 2.0-2.5, 2.6-3.0, 3.1-3.5, 3.6-4.0.
names=csvData.Name
birthmonths= csvData.BirthMonth
plt.plot(birthmonths,names)
plt.gcf().set_size_inches(25,25)
plt.show()
```

```python
#Plotting the HSSC-1 marks of all male vs female students on a scatter plot.
Gender= csvData.Gender
HSSC1_marks= csvData.HSSC1
plt.gcf().set_size_inches(30,15)
plt.scatter(HSSC1_marks,Gender)
```

```python
#Create a correlation matrix between HSSC-1 and HSSC-2 marks and then plot on a heatmap.
sn =csvData.groupby(['HSSC1', 'HSSC2']).size()
csvDataf = pd.DataFrame(sn) 
print(csvDataf)
print(csvData.corr())
ax = sns.heatmap(csvDataf, annot=True)
ax.set(xlabel="", ylabel="")
ax.xaxis.tick_top()
```
